```python
import pandas as pd
df=pd.read_csv("/content/zomato_data.csv")
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.weather.unique()
```

```python
weather_dic ={
    'cloudy': 0 ,
    'rainy': 1,
    'sunny': 2
}
```

```python
df.weather = df.weather.map(weather_dic)
```

```python
df.head()
```

```python
df.time_of_day.unique()
```

```python
time_dic = {
    'peak': 0,
    'off-peak': 1
}
```

```python
df.time_of_day = df.time_of_day.map(time_dic)
```

```python
df.head()
```

```python
df.traffic_level.unique()
```

```python
traffic_dic = {
    'medium': 1 ,
    'high': 2 ,
    'low': 0
}
```

```python
df.traffic_level = df.traffic_level.map(traffic_dic)
```

```python
df.head()
```

```python
df.area_type.unique()
```

```python
area_dic = {
    'urban': 0,
    'suburban': 1,
    'rural': 2
}
```

```python
df.area_type = df.area_type.map(area_dic)
```

```python
df.head(5)
```

```python
df.info()
```

```python
x = df.drop('delivery_time_minutes', axis=1)
y = df.delivery_time_minutes
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```
